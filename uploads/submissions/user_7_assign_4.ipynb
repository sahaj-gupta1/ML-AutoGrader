```python
import pandas as pd

# Load dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
# Basic cleaning
df = df.drop_duplicates()
df = df.fillna(df.median(numeric_only=True))
```

```python
# Separate features and target
features = df.drop(columns=['target'])
target = df['target']
```

```python
from sklearn.model_selection import train_test_split

# Split data
X_tr, X_te, y_tr, y_te = train_test_split(features, target, test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scaling
sc = MinMaxScaler()
X_tr_scaled = sc.fit_transform(X_tr)
X_te_scaled = sc.transform(X_te)
```

```python
from sklearn.linear_model import LogisticRegression

# Model training
model_lr = LogisticRegression(max_iter=200)
model_lr.fit(X_tr_scaled, y_tr)
```

```python
from sklearn.metrics import accuracy_score

# Evaluation
y_predicted = model_lr.predict(X_te_scaled)
print('Model Accuracy:', accuracy_score(y_te, y_predicted))
```
